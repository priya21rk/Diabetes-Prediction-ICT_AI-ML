```python
def remove_outliers(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]

# List of columns to remove outliers from
columns_to_check = [
    'Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness',
    'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']

# Removing outliers for each column
for column in columns_to_check:
    data = remove_outliers(data, column)

print("Data shape after removing outliers:", data.shape)

```

Observations:

i. There are a total of 768 records and 9 features in the dataset.

ii. Each feature can be either of integer or float dataype.

iii. Each column listed has 0 missing values, which means that every record in the dataset is complete for those features.

iv. In the outcome column, 1 represents diabetes positive and 0 represents diabetes negative.

# KNN


```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
metrik_k = []
neighbors = np.arange(3,15)

for k in neighbors:
  classifier = KNeighborsClassifier(n_neighbors=k)
  classifier.fit(x_train,y_train)
  y_pred = classifier.predict(x_test)
  acc = accuracy_score(y_test,y_pred)
  metrik_k.append(acc)
```

```python
metrik_k
```

```python
plt.plot(neighbors,metrik_k,'o-')
plt.xlabel('K value')
plt.ylabel('Accuracy')
plt.grid()
plt.show()
```

```python
classifier = KNeighborsClassifier(n_neighbors=6)
classifier.fit(x_train,y_train)
```

```python
y_pred_knn = classifier.predict(x_test)
accuracy_score(y_test,y_pred_knn)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(x_train,y_train)
y_pred_gnb = gnb.predict(x_test)
accuracy_score(y_test,y_pred_gnb)
```
