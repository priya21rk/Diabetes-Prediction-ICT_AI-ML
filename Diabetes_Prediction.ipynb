## DATA VISUALIZATION

```python
# Visualization 1: Histogram for each of the 9 given features
plt.figure(figsize=(10, 6))
data.hist(bins=30, figsize=(15, 10), layout=(3, 3))
plt.suptitle('Histograms of Numerical Columns', fontsize=16)
plt.show()
```

```python
# Visualization 2: Pie chart for the outcome distribution
import matplotlib.pyplot as plt
import pandas as pd
plt.figure(figsize=(8, 6))
data['Outcome'].value_counts().plot.pie(explode=[0, 0.1], autopct='%1.1f%%', shadow=True, colors=['lightblue', 'salmon'])
plt.title('Outcome Distribution')
plt.ylabel('')
plt.axis('equal')
plt.show()
```

```python
#Visualization 3: Exploring Pregnancy and target variables together
plt.figure(figsize = (8,6))
kde = sns.kdeplot(data["Pregnancies"][data["Outcome"] == 1], color = "lightblue", shade = True)
kde = sns.kdeplot(data["Pregnancies"][data["Outcome"] == 0], ax = kde, color = "salmon", shade= True)
kde.set_xlabel("Pregnancies")
kde.set_ylabel("Density")
kde.legend(["Positive Result", "Negative Result"])
```

```python
#Visualization 4: Correlation matrix graph of the dataset
f, ax = plt.subplots(figsize= [20,15])
sns.heatmap(data.corr(), annot=True, fmt=".2f", ax=ax, cmap = "magma" )
ax.set_title("Correlation Matrix", fontsize=20)
plt.show()
```
